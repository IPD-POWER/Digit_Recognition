```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('zerotonine.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image,cmap="gray")
image.shape
```

Denosing

```python
image = cv2.fastNlMeansDenoising(image,None,10,7,21)
plt.imshow(image,cmap="gray")
```

Thining and skeletonization

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(image,kernel,iterations = 1)
#dilate and erode is oppposite of each other.we are dilating/eroding the backgrounf and not the foreground
plt.imshow(erosion,cmap='gray')
```

adaptive

```python
adapImage = cv2.adaptiveThreshold(erosion,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,19,8)
plt.imshow(adapImage,cmap="gray")
```
